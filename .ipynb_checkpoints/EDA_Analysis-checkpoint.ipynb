```python
#import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#loading data and dropping unnecessary columns usings pandas
data = pd.read_csv('kc_house_data.csv').drop(['yr_renovated','sqft_lot15','sqft_living15', 'zipcode', 'lat', 'long'], axis=1)
data
```

```python
#checking 10 random samples
data.sample(10)
```

```python
# Remove special characters and change data type to float
data['sqft_basement'] = data['sqft_basement'].replace('?', '0').astype(float)
```

```python
data.head()
```

```python
# Check for duplicates in the 'id' column
duplicates = data[data['id'].duplicated()]
duplicates
```

```python
# Remove duplicate 'id' rows, retaining the first occurrence
data = data.drop_duplicates(subset='id', keep='first')
```

```python
#summary of the data
data.info()
```

```python
#checking the number of missing values
sum_missing_values = data.isna().sum()
sum_missing_values
```

```python
#checking for percentage of missing values in relation to total data available
total_data = data.shape[0]
```

```python
# percentage missing value for each column is calculated by dividing sum of each column by len of the whole data
percentage_missing_values = sum_missing_values/total_data *100
#creating a dataframe with the above infor and assign it missing_values_df
missing_values_df =pd.DataFrame({
      'missing values':sum_missing_values,
      'percentage missing values':percentage_missing_values
})
missing_values_df
```

```python
#dropping missing val
data.dropna(subset = ["waterfront", "view"], inplace=True)
```

```python
data.info()
```

```python
#Histograms for numeric features
numeric_features = data.select_dtypes(include=['int64', 'float64']).drop(columns=['id'])
for col in numeric_features.columns:
    plt.figure(figsize=(6, 4))
    plt.hist(data[col], bins=20) # Here a histogram is created  for the current column (col) with 20 bins. It plots the frequency distribution of the values in that column.
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
 #Correlation Analysis
numeric_features = data.select_dtypes(include=['int64', 'float64']).drop(columns=['id'])

correlation_matrix = numeric_features.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
```

```python
#explain the heatmap 
print("""Correlation Coefficient Values:

Positive values like 0.76 and 0.58 have positive linear relationship. As one feature increases, the other tends to increase as well.
Negative values like -0.51 have a negative linear relationship. As one feature increases, the other tends to decrease.
Values close to 0: Features have a weak or no linear relationship.

Strength of Correlation:
Positive correlation: Features move in the same direction. For example, as the number of bedrooms increases, the price of a house tends to increase.
Negative correlation: Features move in opposite directions. For example, as the distance to a city center increases, the price of a house tends to decrease.
No Correlation: Values close to 0 suggest little to no linear relationship between the features.

Diagonal Line: The diagonal line in the heatmap will always be a perfect correlation (correlation coefficient of 1) because it represents each feature's correlation with itself, which is always 1.

Color Intensity: The color intensity in the heatmap represents the strength of the correlation. Darker colors (e.g., dark blue or dark red) indicate stronger correlations, while lighter colors represent weaker correlations.

Use for Feature Selection: Correlation heatmaps can be helpful for feature selection in machine learning. Features that are highly correlated with the target variable (in a positive or negative way) may be important for modeling, while features with low or no correlation may be less informative.

Multicollinearity: High correlations between predictor features (independent variables) can indicate multicollinearity, which can affect the stability and interpretability of regression models.""")
```

```python
#Outlier Detection using boxplot for 'price variable'
plt.figure(figsize=(6, 4))
sns.boxplot(x=data['price'])

plt.title('Boxplot of House Prices')
plt.xlabel('Price')
plt.ylabel('Price Distribution')
plt.show()
```

```python
#explain the outliers
print("""The price column seems to have outliers that are obvious and note but again it is something that can be explained. The difference in the house prices comes about when some features have been considered like say the age of the house can cause the house price to go up or down when a house is very old like say 115 years and it is in bad condition and rated below average are houses with prices below 200,000 while the houses with good conditions, good view, good rating and are also sizeable with morden designs are highly priced.
""")
```

```python
# visualize the House Age calculated by; data["Age"] = data["year_sold"] - data["yr_built"] versus the house prices
plt.figure(figsize=(15, 8))
sns.regplot(x='Age', y='price', data=data, scatter_kws={'alpha':0.5}, color='purple')
plt.title('House Age versus Price')
plt.xlabel('Age')
plt.ylabel('Price')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualizing the relationship between 'sqft_living' and 'price'
sns.scatterplot(data=data, x='sqft_living', y='price')
plt.xlabel('Square Feet of Living Space (sqft)')
plt.ylabel('Price')
plt.title('Relationship between Living Space and Price')
plt.show()
```

```python
# Overlay a regression line on the scatter plot
sns.lmplot(data=data, x='sqft_living', y='price')
plt.xlabel('Square Feet of Living Space (sqft)')
plt.ylabel('Price')
plt.title('Multilinear Relationship between Living Space and Price')
plt.show()
```
